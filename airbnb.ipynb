***Average Airbnb price across countries***

```python
# for index, value in enumerate(price):
#     plt.text(index, value, str(value), ha='center', size=8)

country_price_df = df[['Country', 'Price']] 
country_price_df

grouped = country_price_df.groupby(['Country'])
grouped
avg_price = grouped.mean()
avg_price

# Extracting data for plotting
countries = avg_price.index
average_prices = avg_price['Price']
average_prices = average_prices.sort_values(ascending = False)
bars = plt.bar(countries, average_prices, color='skyblue')

plt.xlabel('Country')
plt.ylabel('Average Price')
plt.title('Average Price per Country')

# Rotating x-axis labels for better readability (optional)
plt.xticks(rotation=90, ha='center')
for index, value in enumerate(average_prices):
    plt.text(index, value, round(value,2), ha='center', size=8, rotation=90)

# Display the plot
plt.show()
```

***Airbnb Host count growth over the years***

```python
df['Host Since'] = pd.to_datetime(df['Host Since'])
df['Host Since']
df.dropna(subset = 'Host Since', inplace=True)
df['Host Since'] = pd.to_datetime(df['Host Since'])
df['Host Since']
df['host_year'] = df['Host Since'].dt.year
df['host_year']
yearly_host_count = df.groupby('host_year')['Host ID'].nunique()
yearly_host_count
```

```python
yearly_host_count.plot(kind='bar', color='skyblue', edgecolor='black')
for index, value in enumerate(yearly_host_count):
    plt.text(index, value, str(value), ha='center')
plt.xticks(rotation=45, ha='center')
plt.xlabel('Year')
plt.ylabel('# of Host IDs')
plt.title('# of Host IDs Created per Year')
plt.tight_layout()
plt.show()
```

***Airbnb Host count growth over the years(in terms of Percentage)***

```python
percentage_growth = yearly_host_count.pct_change() * 100
plt.figure(figsize=(15, 6))
plt.xlabel('Year')
plt.ylabel('Percentage Growth')
plt.title('Percentage Growth of Host IDs over Years')
percentage_growth.plot(kind='line', marker='o', color='green', linestyle='-')
for i, txt in enumerate(percentage_growth):
    plt.annotate(f'{txt:.2f}%', (percentage_growth.index[i], txt),
                 textcoords="offset points", xytext=(0, 5), ha='left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

***Insights***
1. Average price of Airbnb is the **highest in Australia** and the **lowest in Vatican City**.
2. There are around 44 unique property types available on Airbnb. Amongst these property types, the most commonly listed property is **Apartment** with more than **350000** listings. House comes a distant second at **93886** listings.
3. United States has the highest of **134539** listings.
4. In terms of Host Response Rate, both **Mexico** and **Vatican City** top the list at **100%** while **China** has the lowest Host Response Rate at **88.87%**. However, it has to be noted that, both **Mexico** and **Vatican City** each have only 2 listings on Airbnb.
5. There are 3 types of room listed - *Entire home/apt, Shared room and Private room*. Amongst these, the room type with the maximum percentage of listing is **Entire home/apt** with a share of **63.38%**. Shared room has the least percentage of listings with a **1.81%** share.
6. Airbnb had the maximum number of hosts added in the year **2015** when more than **90000** new Host ID's were created. However, since then, there has been a downward trend with a lesser number of new Host ID's being created every year.
7. In terms of percentage, maximum increase was of **635.37%** in the year **2009** when the Host count increased from 147 to 1081. However, since the year 2016 there has been a steady decline in the user count growth. In fact, there has been a ***negative growth*** from 2016 to 2017 wherein host count percenatge reduced from **-24.10% to -80.34%** or in terms of numbers, in 2016 about **68416** new hosts were added whereas in 2017 there was an addition of only **13454** new Hosts.
8. It has been infered from the dataset that throughtout the world, there isn't much of a drastic change in the size of the property offered by the host. On an average, the property size is around **575 square feet**. The highest average is found in China (591 sq. feet; however there were only two observations. The lowest was in Italy (567 sq. feet))
9. In terms of **Cancellation Policy**, most of the Airbnb's across the globe have a **strict** policy(182641), folowed by **flexible**(156169) and **moderate**(122132).
10. Although the property size across all the countries is almost the same, the accomodatio capacity varies quite a bit. For instance, even though the United States have an average Property Size of 581.35 sq. feet, they allow accomodation for approximately 3 people on an average. Also, **Vanuatu**, has the lowest accomodation average amongst all(1)(Only one property have been listed for Vanuatu in the dataset; so the insight might be a little skewed).
